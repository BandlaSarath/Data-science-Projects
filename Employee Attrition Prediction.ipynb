Label Encoding: converting categorical data into numerical data

```python
from sklearn.preprocessing import LabelEncoder
label_encoder=LabelEncoder()
```

```python
emp_data['salary']=label_encoder.fit_transform(emp_data['salary'])
emp_data['Department']=label_encoder.fit_transform(emp_data['Department'])
```

```python
emp_data.head()
```

**Dependent and independent variable:**


```python
x=emp_data.drop('left',axis=1)
y=emp_data['left']
```

**Dividing the dataset into train and test**

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
std_scaler=StandardScaler()
X_train_scaled =std_scaler.fit_transform(X_train)
X_test_scaled =std_scaler.fit_transform(X_test)
```

**Machine learning model**--Support vector machine






```python
from sklearn.svm import SVC
```

```python
# Create and train an SVM classifier
svm_classifier = SVC(kernel='linear')  # You can also try other kernels like 'rbf', 'poly', etc.
svm_classifier.fit(X_train_scaled, y_train)
```

```python
y_pred = svm_classifier.predict(X_test_scaled)
```

**Accuracy score**

```python
from sklearn.metrics import accuracy_score
```

```python
# Evaluate the classifier
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

**Classification Report**

```python
from sklearn.metrics import classification_report

# Assuming 'y_test' contains the true labels and 'y_pred' contains the predicted labels
# Generate a classification report
report = classification_report(y_test, y_pred)

# Print the classification report
print(report)
```
